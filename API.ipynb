```python
import requests
import datetime
import json
import pandas as pd
import re
import pickle
```

```python
#df_presenters = pd.read_pickle("Files/Dataframes/Presenters_wGender.pkl")  #with gender
df_presenters = pd.read_pickle("Files/Dataframes/Presenters.pkl") #all presenters(wihtout gender)
#instead load new df after genderEstimation
```

```python
authors = list(df_presenters.Name.unique())
```

```python
#create authorID, map to the author by creating dictionary and save
authorID = {i: authors[i] for i in range(len(authors))}

#save dict
with open('Files/Dictionary/authorID.pkl', 'wb') as fp:
    pickle.dump(authorID, fp)
```

```python
def saveAuthor(names, path, allNames):
    for i in names:
        name = re.sub(' ', '%20', i)
        response = requests.get("https://api.openalex.org/authors?search=" + name)
        author = response.json()
        index = allNames.index(i)
        with open(path + str(index) + '.txt', "w") as f:
            json.dump(author, f)
```

```python
#first run. author 0-14507
saveAuthor(authors, 'Files/Authors/', authors)
```
